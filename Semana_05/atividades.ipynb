```python
#Importe os arquivos clientes.csv, pedidos.csv e produtos.csv usando Pandas. 
#Explore os dados para entender sua estrutura.

%pip install --upgrade pandas
import pandas as pd

clientes = pd.read_csv('clientes.csv')
pedidos = pd.read_csv('pedidos.csv')
produtos = pd.read_csv('produtos.csv')

print("Clientes:")
print(clientes.head())  

print("Pedidos:")
print(pedidos.head())

print("Produtos:")
print(produtos.head())

print("Estrutura dos dados dos clientes:")
print(clientes.info())
print("Estrutura dos dados dos pedidos:")
print(pedidos.info())   
print("Estrutura dos dados dos produtos:")
print(produtos.info())

print("Descrição estatística dos dados dos clientes:")
print(clientes.describe(include='all'))
print("Descrição estatística dos dados dos pedidos:")
print(pedidos.describe(include='all'))
print("Descrição estatística dos dados dos produtos:")
print(produtos.describe(include='all'))
```

```python
#Aplique filtros para extrair informações específicas dos DataFrames. 
#Selecionar clientes com idade > 30. Filtrar pedidos com valor > R$500 
#Selecionar produtos da categoria "Eletrônicos". Usar .loc[] e olic[] para seleções

clientes_maiores_30 = clientes.loc[clientes['idade'] > 30]
print("Clientes com idade > 30:")
print(clientes_maiores_30)
pedidos_valor_acima_500 = pedidos.loc[pedidos['valor_total'] > 500]
print("Pedidos com valor > R$500:")
print(pedidos_valor_acima_500)
produtos_eletronicos = produtos.loc[produtos['categoria'] == 'Eletrônicos']
print("Produtos da categoria 'Eletrônicos':")
print(produtos_eletronicos)
```

```python
#Realize operações para cruzar e resumir os dados. 
#Agrupar pedidos por cliente e somar valores (groupby) 
#Juntar clientes e pedidos (merge) 
#Criar tabela dinâmica com categorias e média de preços (pivot_table) 
#Verificar duplicatas e limpar dados se necessário

resumo_pedidos = pedidos.groupby("id_cliente")["valor_total"].sum().reset_index()

clientes_pedidos = pd.merge(clientes, resumo_pedidos, on="id_cliente", how="left")

tabela_categorias = produtos.pivot_table(values="preco", index="categoria", aggfunc="mean")

clientes.duplicated().sum()
pedidos.duplicated().sum()
produtos.duplicated().sum()
```

```python
#Utilize NumPy para cálculos e simulações com arrays. 
#Importar NumPy como np Criar array com valores dos pedidos 
#Calcular média, soma e desvio padrão 
#Aplicar desconto de 10% com broadcasting 
#Verificar tipos e dimensões com .dtype() e .shape()

import numpy as np

# Criar um array com os valores dos pedidos (exemplo)
valores_pedidos = np.array([120.50, 85.90, 230.00, 99.99, 175.30])

# Exibir o array original
print("Valores dos pedidos:", valores_pedidos)

# Calcular a soma dos pedidos
soma = np.sum(valores_pedidos)
print("\nSoma total dos pedidos: R$", soma)

# Calcular a média dos pedidos
media = np.mean(valores_pedidos)
print("Média dos pedidos: R$", media)

# Calcular o desvio padrão dos pedidos
desvio_padrao = np.std(valores_pedidos)
print("Desvio padrão dos pedidos: R$", round(desvio_padrao, 2))

# Aplicar desconto de 10% usando broadcasting
desconto = valores_pedidos * 0.90
print("\nValores com 10% de desconto:", desconto)

# Verificar tipo de dados (dtype)
print("\nTipo de dados do array:", valores_pedidos.dtype)

# Verificar formato/dimensões (shape)
print("Dimensões do array:", valores_pedidos.shape)
```

```python
#Produza um resumo com os principais insights da análise. 
#Identificar o cliente que mais gastou 
#Calcular ticket médio por categoria 
#Verificar cidade com maior número de clientes 
#Criar visualizações (opcional)

import pandas as pd
import numpy as np

#Importar os arquivos CSV (exemplo)
clientes = pd.read_csv("clientes.csv")
pedidos = pd.read_csv("pedidos.csv")
produtos = pd.read_csv("produtos.csv")

#Exibir as primeiras linhas para conferência
print("Clientes:\n", clientes.head(), "\n")
print("Pedidos:\n", pedidos.head(), "\n")
print("Produtos:\n", produtos.head(), "\n")

#Cruzar e resumir os dados

#Juntar pedidos com clientes e produtos
dados_completos = pedidos.merge(clientes, on="id_cliente").merge(produtos, on="id_produto")

#Exibir dataset combinado
print("Dados combinados:\n", dados_completos.head(), "\n")

#INSIGHTS DA ANÁLISE

#Cliente que mais gastou
gasto_por_cliente = dados_completos.groupby("nome_cliente")["valor_total"].sum()
cliente_top = gasto_por_cliente.idxmax()
valor_top = gasto_por_cliente.max()
print(f"Cliente que mais gastou: {cliente_top} (R$ {valor_top:.2f})")

#Ticket médio por categoria
ticket_categoria = dados_completos.groupby("categoria")["valor_total"].mean()
print("\nTicket médio por categoria:\n", ticket_categoria)

#Cidade com maior número de clientes
clientes_por_cidade = clientes["cidade"].value_counts()
cidade_top = clientes_por_cidade.idxmax()
qtde_top = clientes_por_cidade.max()
print(f"\nCidade com maior número de clientes: {cidade_top} ({qtde_top} clientes)")

#VISUALIZAÇÕES

import matplotlib.pyplot as plt

#Gráfico 1 - Top 5 clientes que mais gastaram
gasto_por_cliente.sort_values(ascending=False).head(5).plot(kind="bar", title="Top 5 Clientes que Mais Gastaram", color="skyblue")
plt.xlabel("Cliente")
plt.ylabel("Valor Total (R$)")
plt.show()

#Gráfico 2 - Ticket médio por categoria
ticket_categoria.plot(kind="barh", title="Ticket Médio por Categoria", color="orange")
plt.xlabel("Valor Médio (R$)")
plt.ylabel("Categoria")
plt.show()

#RESUMO FINAL

print("\n📊 RESUMO GERAL:")
print(f"- Cliente que mais gastou: {cliente_top} (R$ {valor_top:.2f})")
print(f"- Cidade com mais clientes: {cidade_top} ({qtde_top} clientes)")
print("\nTicket médio por categoria:\n", ticket_categoria)
```
